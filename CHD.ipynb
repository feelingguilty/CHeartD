```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df['education'].fillna(df['education'].mean(),inplace=True)
```

```python
df['cigsPerDay'].fillna(df['cigsPerDay'].mean(),inplace=True)
```

```python
df['BPMeds'].fillna(df['BPMeds'].mean(),inplace=True)
df['totChol'].fillna(df['totChol'].mean(),inplace=True)
df['BMI'].fillna(df['BMI'].mean(),inplace=True)
df['heartRate'].fillna(df['heartRate'].mean(),inplace=True)
df['glucose'].fillna(df['glucose'].mean(),inplace=True)
```

```python
x = df.drop(columns = 'TenYearCHD')
y = df['TenYearCHD']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.head()
```

```python
for col in x_train.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(x_train[col])
    plt.title(col)

```

```python
df.info()
```

```python
imputer = SimpleImputer(strategy='mean')
ct = ColumnTransformer(transformers=[('tnf1', imputer,['cigsPerDay', 'BPMeds', 'totChol', 'BMI', 'heartRate', 'glucose'])],remainder='passthrough')
```

```python
newx_train = ct.fit_transform(x_train)
newx_test = ct.transform(x_test)
```

```python
scaler = StandardScaler()
scaler.fit(newx_train)
newx_train_scaled = scaler.transform(newx_train)
newx_test_scaled = scaler.transform(newx_test)

lr = LogisticRegression()
dt = DecisionTreeClassifier()

lr.fit(newx_train_scaled,y_train)
dt.fit(newx_train_scaled,y_train)
```

```python
lr_pred = lr.predict(newx_test_scaled)
dt_pred = dt.predict(newx_test_scaled)
```

```python
print("Logistic Regression:",accuracy_score(y_test,lr_pred)*100)
print('Decision Tree:',accuracy_score(y_test,dt_pred)*100)
```

```python
minmax = MinMaxScaler()
minmax = minmax.fit(newx_train)

mx_train_scaled = minmax.transform(newx_train)
mx_test_scaled = minmax.transform(newx_test)


lr.fit(newx_train_scaled,y_train)
dt.fit(newx_train_scaled,y_train)
```

```python
lr_pred_mx = lr.predict(mx_test_scaled)
dt_pred_mx = dt.predict(mx_test_scaled)
```

Using MinMax Scaler


```python
print('MinMax Scaler-Logistic Regression:',accuracy_score(y_test,lr_pred_mx)*100)
print('MinMax Scaler-Decision Tree:',accuracy_score(y_test,dt_pred_mx)*100)
```
